```python
import pandas as pd,numpy as np
df=pd.read_csv('alphabet_stock_data .csv')
import matplotlib.pyplot as plt
import os
```

```python
#Write a Pandas program to create a line plot of the opening, closing stock prices of Alphabet Inc. between two specific dates
df=pd.DataFrame(df,columns=['Date','Close'])
print(df)
filterr=np.where((df.Date>='2020-04-01') & (df.Date<='2020-04-07'))
var=df.iloc[filterr[0]]
print(var)
var.plot(x='Date',y='Close',color='r')
#var.index=var.Date
#var.plot()
```

```python
#Write a Pandas program to create a line plot of the historical stock prices of Alphabet Inc. between two specific dates.
filterr=np.where((df.Date>='2020-04-01') & (df.Date<='2020-04-15'))
var=df.iloc[filterr[0]]
print(var)
var.plot(x='Date',y='Close',color='b')
```

```python
df=pd.DataFrame(df,columns=['Date','Volume'])
print(df)
filterr=np.where((df.Date>='2020-04-01') & (df.Date<='2020-04-15'))
var=df.iloc[filterr[0]]
print(var)
var.plot.bar(x='Date',y='Volume',color='b')
```

```python
# Write a Pandas program to create a bar plot of opening,closing stock prices of Alphabet Inc. between two specific dates
df=pd.DataFrame(df,columns=['Date','Close','Open'])
print(df)
filterr=np.where((df.Date>='2020-04-01') & (df.Date<='2020-04-15'))
var=df.iloc[filterr[0]]
print(var)
var.plot.bar(x='Date',y=['Close','Open'])
```
