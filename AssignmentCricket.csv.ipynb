```python
df=pd.read_csv('Cricket.csv',encoding= 'unicode_escape')
```

```python
df.head(27)
```

```python
new_df=df[['SR','Ave']]
```

```python
new_df.head()
```

```python
km_mod=KMeans(n_clusters=4)
```

```python
km_mod=km_mod.fit(new_df)
```
